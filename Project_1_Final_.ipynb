# A-Imputation of Numeric Features with Median

1) Create a list of features

```python
# Get the list of numeric column names
num_cols = X_train.select_dtypes("number").columns
num_cols
```

2) Instantiate the Imputer

```python
from sklearn.impute import SimpleImputer
```

```python
# Instantiate the imputer object from the SimpleImputer class with strategy 'median'
impute_median = SimpleImputer(strategy='median')
```

3) Fit the Imputer of the Training Data

```python
# Fit the imputer object on the numeric training data with .fit()
impute_median.fit(X_train[num_cols])
```

4a) Transform the Training Data

```python
from sklearn import set_config
set_config(transform_output='pandas')
```

```python
# Transform the training data (fill in nulls with median)
X_train_num_imputed = impute_median.transform(X_train[num_cols])
X_train_num_imputed
```

```python
#Let's confirm that our transformed numeric training data no longer has null values.
X_train_num_imputed.isna().sum()
```

4b) Transform the Test Data

```python
# Transform the testing data (fill in nulls with median from training data)
X_test_num_imputed = impute_median.transform(X_test[num_cols])
X_test_num_imputed.head()
```

```python
#Let's confirm that our transformed numeric training data no longer has null values.
X_test_num_imputed.isna().sum()
```

# FINAL PART

1. Build a linear regression model to predict sales.

* Build a linear regression model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg
```

```python
# Fit the model on the training data
lin_reg.fit(X_train_processed, y_train)
```

```python
# Get predictions for the training data
y_predictions_train = lin_reg.predict(X_train_processed)
# Get predictions for the testing data
y_predictions_test = lin_reg.predict(X_test_processed)
```

* Evaluate the performance of your model based on r^2

```python
from sklearn.metrics import r2_score
train_r2 = r2_score(y_train, y_predictions_train)
test_r2 = r2_score(y_test, y_predictions_test)
print(f'Model Training R2: {train_r2:.2f}')
print(f'Model Testing R2: {test_r2:.2f}')
```

* Evaluate the performance of your model based on rmse.

```python
from sklearn.metrics import mean_squared_error
train_RMSE = mean_squared_error(y_train, y_predictions_train, squared=False)
test_RMSE = mean_squared_error(y_test, y_predictions_test, squared=False)
print(f'Model Training RMSE: {train_RMSE:,.2f}')
print(f'Model Testing RMSE: {test_RMSE:,.2f}')
```

2. Your second task is to build a regression tree model to predict sales.

* Build a simple regression tree model

```python
from sklearn.tree import DecisionTreeRegressor
# Make an instance of the model
dec_tree = DecisionTreeRegressor(random_state = 42)
# Training the Model on the data
dec_tree.fit(X_train_processed, y_train)
# Make predictions for training and testing data
train_preds_dec = dec_tree.predict(X_train_processed)
test_preds_dec = dec_tree.predict(X_test_processed)
```

* Compare the performance of your model based on r^2.

```python
train_score = r2_score(y_train, train_preds_dec)
test_score = r2_score(y_test, test_preds_dec)
print (f'Training r2 is {train_score:.2f}.')
print (f'Testing r2 is {test_score:.2f}.')
```

* Compare the performance of your model based on rmse.

```python
train_RMSE_dec = mean_squared_error(y_train, train_preds_dec,squared=False)
test_RMSE_dec = mean_squared_error(y_test, test_preds_dec,squared=False)
print (f'Training RMSE is {train_RMSE_dec:.2f}.')
print (f'Testing RMSE is {test_RMSE_dec:.2f}.')
```
